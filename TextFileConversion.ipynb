```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv('/content/text.csv')
df
x = df['FILE SIZE']
y1 = df['C']
y2 = df['C++']
y3 = df['JAVA']
y4 = df['R']
y5 = df['PYTHON']
pic = sns.lineplot(x=x,y=y1, label = 'C')
sns.lineplot(x=x,y=y2, label = 'C++')
sns.lineplot(x=x,y=y3, label = 'JAVA')
sns.lineplot(x=x,y=y4, label = 'R')
sns.lineplot(x=x,y=y5, label = 'PYTHON')
pic.set_ylabel('Time taken (in sec)')
pic.set_xlabel('File size')
plt.savefig("IMAGE.jpg", dpi = 300)

```
